```python
import os 
import cv2
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
img=cv2.imread(r"C:\Users\rauna\Desktop\WhatsApp Image 2024-05-23 at 12.52.14.jpeg")
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
lower = np.array([110,110,110])
higher=np.array([250,250,250])
```

```python
mask=cv2.inRange(img,lower,higher)
```

```python
mask.shape
```

```python
plt.imshow(mask)
```

```python
cont,_=cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
cont_img=cv2.drawContours(img,cont,-1,255,5)
```

```python
plt.imshow(cont_img)
```

```python
c=max(cont,key=cv2.contourArea)
```

```python
c
```

```python
x,y,w,h=cv2.boundingRect(c)
```

```python
cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
plt.imshow(img)
```

```python
cropped_image=img[y:y+h,x:x+w]
 
```

```python
plt.imshow(cropped_image)
```
